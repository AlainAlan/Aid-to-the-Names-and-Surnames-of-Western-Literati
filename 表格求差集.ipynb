```python
import pandas as pd
```

```python
hhz = pd.read_excel('HHZ.xlsx')
pc = pd.read_excel('PC2021.xlsx')

print(type(hhz))
print(hhz)
print(pc)
```

```python
合集 = hhz.append(pc,ignore_index=True)
合集.shape
```

```python
合集重复行 = 合集.duplicated(['汉名'],keep=False)
合集重复数据 = 合集.loc[合集重复行,:]
```

```python
HHZ加上重复数据 = hhz.append(合集重复数据,ignore_index=True)
PC加上重复数据 = pc.append(合集重复数据,ignore_index=True)
```

```python
HHZ无共有行 = HHZ加上重复数据.duplicated(['汉名'],keep=False)
PC无共有行 = PC加上重复数据.duplicated(['汉名'],keep=False)
```

```python
# 合集無重複 = 合集.drop_duplicates(['汉名'],keep=False)
# 合集無重複.shape
```

```python
HHZ獨有部分 = HHZ加上重复数据.drop_duplicates(['汉名'],keep=False)
PC独有部分 = PC加上重复数据.drop_duplicates(['汉名'],keep=False)
```
